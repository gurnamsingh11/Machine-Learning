## Diabetes Predictions

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes

dataset = load_diabetes()
```

```python
print(dataset.DESCR)
```

```python
print(dataset)
```

```python
df_diabetes = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df_diabetes.head()
```

```python
X = df_diabetes
y = dataset.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
X_train.head()
```

### Correlation

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

### Hyperparameter Tuning

```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,20,25],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor, param_grid=param, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings("ignore")

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

### Visualize the Decision Tree

```python
selected_model = DecisionTreeRegressor(criterion='squared_error', max_depth=4, max_features='sqrt', splitter='random')
```

```python
selected_model.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected_model, filled=True)
```
