## Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("height-weight.csv")
df.head()
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
```

### Divide dataset into Independent and Dependent Features

```python
X = df[["Weight"]]
y = df["Height"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape, X_test.shape
```

### Standardize train independent data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
plt.scatter(X_train, y_train)
```

### Train the Model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
print("Slope or Coeffiecient of Weight is", regressor.coef_) # theta 1
print("Intercept: ", regressor.intercept_) # theta 0
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regressor.predict(X_train), "r")
```

### Prediction of train data
1. Predicted height output = intercept + coef_(Weights)
2. y_pred_train = 157.5 + 17.03(X_train)

### Prediction of test data

1. prediction height output = intercept + coef_(Weights)
2. y_pred_test = 157.5 + 17.03(X_test)

```python
y_pred_test = regressor.predict(X_test)
```

```python
y_pred_test, y_test # predicted and actual
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, regressor.predict(X_test), 'r')
```
